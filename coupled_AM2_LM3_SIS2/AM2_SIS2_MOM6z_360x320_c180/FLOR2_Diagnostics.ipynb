This suite of experiments are documented in this [XML](https://github.com/NOAA-GFDL/MOM6-examples/blob/user/mjh/FLOR2/ice_ocean_SIS2/FLOR2/xml/gaea/FLOR2_LM3_SIS2.xml).  
<p>

All experiments initialized from rest using January WOA05 temperature and salinity climatology.

Along tracer mixing and thickness diffusion are currently disabled.

<I>Sensitivity test for vertical diffisivities and mixed layer dynamics.</I>
<p>
<B>Baseline1:</B> Default settings for EPBL and background (AJA).
<B>MS03:</B> MSTAR=0.3 (1.2 in Baseline1) MSTAR is a non-dimensional efficiency for mechanical production of turbulence by winds.
<B>MS03_noBg:</B> Same as MS03 but with background diffusivities effectively set to zero.
<B>iceRes:</B> initialized ice model with Jan 1, 1983 ice state from Baseline1.

The total heat uptake by the model from 1980-1987 exceeds expected values, ~3.3-4.5 W m-2. MS03_noBg, which has reduced mechanical stirring of the mixed layer, has the lowest uptake. MSTAR has negligible impact on average heat uptake, but does impact the seasonal cycle of heat ventilation in the ocean. Background dissipation increases heat uptake by ~30%, mostly due to increase during Boreal winter mixing.

Looking regionally, in the North Atlantic and Arctic, the ocean releases heat to the atmosphere during winter months, up to -3 PWatts, or more than -100 W m-2, during convection.  The ocean gains back most of this lost heat in the summer months, and on average, the North Atlantic and Arctic lose about 0.5 PWatts, or -20 W m-2.  Additional mixing, either from increasing boundary layer mechanical turbulence, or from prescribed background diffusivities increases the amplitude of the seasonal cycle in the North Atlantic.
```python
from midas.rectgrid import *
import numpy as np
import matplotlib.pyplot as plt
import netCDF4 as nc
import hashlib
import glob

sgrid=supergrid(file='ocean_hgrid.nc',tripolar_n=True,cyclic_x=True)
grid=quadmesh(supergrid=sgrid)
grid.D=nc.Dataset('topog.nc').variables['depth'][:]
grid.wet=np.zeros(grid.D.shape)
grid.wet[grid.D>0.]=1.
grid.lath=grid.y_T[:,grid.im/4]
mask=nc.Dataset('basin_codes.nc').variables['basin'][:]
AtlMask=mask.copy()
AtlMask[mask==1]=0.
AtlMask[mask==2]=1.
AtlMask[mask==4]=1.
AtlMask[mask==6]=1.
AtlMask[mask==8]=1.
AtlMask[mask==9]=1.
AtlMask[AtlMask>1]=0.
PacMask=mask.copy()
PacMask[mask!=3]=0.
PacMask[mask==3]=1.
grid.mask=grid.wet.copy()

EXP=[]
EXP.append({'name':'FLOR2_LM3_SIS2_baseline1',\
            'root_path':'/ptmp/Matthew.Harrison/archive/Matthew.Harrison/ulm_mom6_d00d00100e7cc79130abeadbbda1997f698f6659/',\
            'platform':'gfdl.ncrc2-intel-prod','grid':grid})
EXP.append({'name':'FLOR2_LM3_SIS2_baseline1_MS03',\
            'root_path':'/ptmp/Matthew.Harrison/archive/Matthew.Harrison/ulm_mom6_d00d00100e7cc79130abeadbbda1997f698f6659/',\
            'platform':'gfdl.ncrc2-intel-prod','grid':grid})
EXP.append({'name':'FLOR2_LM3_SIS2_baseline1_MS03_noBg',\
            'root_path':'/ptmp/Matthew.Harrison/archive/Matthew.Harrison/ulm_mom6_d00d00100e7cc79130abeadbbda1997f698f6659/',\
            'platform':'gfdl.ncrc2-intel-prod','grid':grid})
EXP.append({'name':'FLOR2_LM3_SIS2_baseline1_iceRes',\
            'root_path':'/ptmp/Matthew.Harrison/archive/Matthew.Harrison/ulm_mom6_d00d00100e7cc79130abeadbbda1997f698f6659/',\
            'platform':'gfdl.ncrc2-intel-prod','grid':grid})



def show_variables(filename):
    f=nc.Dataset(filename)
    V=[]
    for v in f.variables:
        print v
        V.append(v)
    return V

def get_listing(exp,suffix):
    listing=exp['root_path']+exp['name']+'/'+exp['platform']+suffix
    listing=sorted(glob.glob(listing))
    return listing
    
def get_model_output(EXP=None,suffix=None,fields=None):

    S=[]
    for exp in EXP:
        listing=get_listing(exp,suffix)
        S.append(state(MFpath=listing,grid=exp['grid'],\
                   fields=fields,verbose=False,\
                   date_bounds=date_bounds,\
                   default_calendar='gregorian'))
    
    return S


def plot_sfc_xyav(S,EXP,field=None,figsize=None,scale=1.0,mask=None,title=None,normalize=True):
    fig=plt.figure(1,figsize=figsize)
    ax1=fig.add_subplot(111)
#    ax2=fig.add_subplot(312)
#    ax3=fig.add_subplot(313)

    if normalize==True:
        suffix='_xyav'
    else:
        suffix='_xyint'
    colList=['b','r','g','c','k','y'][::-1]
    for s,exp in zip(S,EXP):
        col=colList.pop()
        dates=instance_to_datetime(s.var_dict[field]['dates'])
        s.unmask(field)
        if mask is not None:
            s.grid.mask=mask
        else:
            s.grid.mask=s.grid.wet
        s.mask_where(field,'grid.mask==0')
        s.volume_integral(field,'XY',normalize=normalize)
        res=scale*sq(eval('s.'+field+suffix))
        s.time_avg(field+suffix)
        cres=str(scale*eval('sq(s.'+field+suffix+'_tav)'))[0:6]
        txt=exp['name']+' ('+cres+')'
        ax1.plot_date(dates,res,color=col,linestyle='solid',label=txt)
    ax1.grid()
    ax1.set_title(title)
    ax1.legend(prop={'size':8})

    return fig

def plot_implied_meridional_transport(S,EXP,field=None,figsize=None,scale=1.0,mask=None,\
                                      title=None,obs_ref=None,caxis=None,obs_label=None):

    fig=plt.figure(1,figsize=figsize)
    ax1=fig.add_subplot(111)

    colList=['b','r','g','c','k','y'][::-1]
    x=S[0].grid.lath

    for s,exp in zip(S,EXP):
        col=colList.pop()
#    print col,exp['name']
        s.unmask(field)
        if mask is not None:
            s.grid.mask=mask
        else:
            s.grid.mask=s.grid.wet
        s.mask_where(field,'grid.mask==0.')
        s.time_avg(field)
        s.volume_integral(field+'_tav','X',normalize=False)
        y=eval('s.'+field+'_tav_xint')
        y=sq(np.ma.filled(y,0.))
        y=np.sum(y)-np.cumsum(y)
        ax1.plot(x,y*scale,color=col,label=exp['name'])

    if obs_ref is not None:
        ax1.plot(obs_ref[0][0],obs_ref[0][0],obs_ref[0][1],obs_ref[0][2],color='k',linestyle='dashed',label=obs_label)
        ax1.plot(obs_ref[1][0],obs_ref[1][0],obs_ref[1][1],obs_ref[1][2],color='k',linestyle='dashed',label=obs_label)

    if caxis is not None:
        ax1.set_xlim(caxis[0],caxis[-1])
#ax1.set_ylim(-1.0,2.0)
    ax1.legend(loc='upper left')
    ax1.grid()
    ax1.set_title(title)
    
    return fig

def plot_meridional_overturning(S,EXP,field='vh',figsize=None,scale=1.0,mask=None,\
                                y_range=None,title=None,obs_label=None):

    fig=plt.figure(1,figsize=figsize)
    ax1=fig.add_subplot(111)

    colList=['b','r','g','c','k','y'][::-1]
    x=S[0].grid.lath


    indx=np.where(np.logical_and(x>y_range[0],x<y_range[1]))
#    print indx
    for s,exp in zip(S,EXP):
        dates=instance_to_datetime(s.var_dict[field]['dates'])
        col=colList.pop()
        s.unmask(field)
        if mask is not None:
            s.grid.mask=mask
        else:
            s.grid.mask=s.grid.wet
        s.mask_where(field,'grid.mask==0.')
        y=eval('s.'+field)
        y=np.ma.filled(y,0.)
        y=np.ma.sum(y,axis=3)
        y=np.cumsum(y,axis=1)*scale
        y=sq(y[:,:,indx])
        y=np.max(y,axis=2)
        y=np.max(y,axis=1)
        ax1.plot_date(dates,y,linestyle='solid',color=col,label=exp['name'])
        

    ax1.legend(loc='lower left')
    ax1.grid()
    ax1.set_title(title)
    
    return fig

def plot_MOC(s,exp,field='vh',figsize=None,scale=1.0e-9,mask=None,\
                                y_range=None,fill_levs=[-25.,25.],clevs=np.arange(-30,35,5),title=None,cmap=plt.cm.bwr):

    fig=plt.figure(1,figsize=figsize)
    ax1=fig.add_subplot(111)

    x=s.grid.lath
    y=s.var_dict[field]['zax_data']

    X,Y=np.meshgrid(x,y)

    s.time_avg(field)
    z=sq(np.ma.filled(eval('s.'+field+'_tav'),0.))
    z=np.ma.sum(z,axis=2)
    z=np.cumsum(z,axis=0)
    z=np.ma.masked_where(z==0.,z)
    cf=ax1.pcolormesh(X,-Y,z/1.e9,vmin=fill_levs[0],vmax=fill_levs[-1],cmap=cmap)
    cs=ax1.contour(X,-Y,z/1.e9,clevs,colors='k')
    plt.clabel(cs,format='%d3')
    ax1.set_xlim(y_range[0],y_range[1])
    ax1.set_ylim(-5500,0)
    ax1.grid()
    plt.colorbar(cf,ax=ax1)
    ax1.set_title(title)
    
    return fig

def plot_xav(s,exp,field=None,figsize=None,scale=1.0,mask=None,\
        y_range=None,z_range=None,fill_levs=None,clevs=None,title=None,cmap=plt.cm.spectral):

    fig=plt.figure(1,figsize=figsize)
    ax1=fig.add_subplot(111)


    s.unmask(field)
    s.grid.mask = mask
    s.mask_where(field,'grid.mask==0')
    x=s.grid.lath
    y=s.var_dict[field]['zax_data']
    X,y=np.meshgrid(x,y)
    s.volume_integral(field,'X',normalize=True)
    z=sq(eval('s.'+field+'_xav[-1]')-eval('s.'+field+'_xav[0]'))
    Y=sq(eval('s.'+'var_dict[\''+field+'_xav\'][\'z\']'))
    cf=ax1.pcolormesh(X,Y,z,vmin=fill_levs[0],vmax=fill_levs[-1],cmap=cmap)
    ax1.contour(X,Y,z,clevs,colors='k')
    ax1.set_xlim(y_range[0],y_range[1])
    ax1.set_ylim(z_range[0],z_range[1])
    ax1.grid()
    plt.colorbar(cf,ax=ax1)
    ax1.set_title(title+str(s.var_dict[field]['dates'][-1]),fontsize=7)
```
