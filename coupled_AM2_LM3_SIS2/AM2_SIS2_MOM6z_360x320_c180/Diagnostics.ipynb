```python
O=[]

date_bounds=[datetime.datetime(10,1,1),datetime.datetime(10,12,31)]


for exp in EXP:
    exp['geo_region']=None
    
for exp in EXP:    
    listing=exp['root_path']+exp['name']+'/'+exp['platform']+\
    '/history/*/*.ocean_month.nc'
#    print listing
    listing=glob.glob(listing)
    listing=sorted(listing)
    O.append(state(MFpath=listing,grid=grid,geo_region=exp['geo_region'],\
                   fields=['temp'],verbose=False,\
                   default_calendar='gregorian',date_bounds=date_bounds))


for o in O:
    o.adjust_thickness('temp')
    o.var_dict['temp']['Ztype']='Fixed'
    o.time_avg('temp')
    o.volume_integral('temp_tav','X',normalize=True)
    o.add_interface_bounds('temp_tav_xav')
```

```python
%matplotlib inline

fig=plt.figure(1,figsize=(14,8))
ax1=fig.add_subplot(111)
xax=O[0].grid.latq
YAX=sq(O[0].var_dict['temp_tav_xav']['z_interfaces_ns'])
YAX[YAX<-1.e4]=0.
yax=np.arange(YAX.shape[0])
XAX,ydum=np.meshgrid(xax,yax)
zout=sq(O[1].temp_tav_xav-O[0].temp_tav_xav)
cf=ax1.pcolormesh(XAX,YAX,zout,vmin=-2,vmax=2,cmap=cmap_bwr)
ax1.set_ylim(-1500,0)
plt.xlabel('Latitude (degN)')
plt.ylabel('Depth (meters)')
tit='Zonal Average Temperature Anomaly (degC) year '+str(O[0].var_dict['temp_tav']['dates'].year)
plt.title(tit,fontsize=10)
plt.colorbar(cf)
plt.grid()
```
