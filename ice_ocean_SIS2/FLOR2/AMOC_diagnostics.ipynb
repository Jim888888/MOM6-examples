```python
EXP=[]
EXP.append({'name':'FLOR2_LM3_SIS2_baseline0',\
            'root_path':'/ptmp/Matthew.Harrison/archive/Matthew.Harrison/ulm_mom6_d00d00100e7cc79130abeadbbda1997f698f6659/',\
            'platform':'gfdl.ncrc2-intel-prod'})
EXP.append({'name':'FLOR2_LM3_SIS2_baseline0_KPP',\
            'root_path':'/ptmp/Matthew.Harrison/archive/Matthew.Harrison/ulm_mom6_d00d00100e7cc79130abeadbbda1997f698f6659/',\
            'platform':'gfdl.ncrc2-intel-prod'})
EXP.append({'name':'FLOR2_LM3_SIS2_baseline0_no_rivers',\
            'root_path':'/ptmp/Matthew.Harrison/archive/Matthew.Harrison/ulm_mom6_d00d00100e7cc79130abeadbbda1997f698f6659/',\
            'platform':'gfdl.ncrc2-intel-prod'})
EXP.append({'name':'FLOR2_LM3_SIS2_baseline0_layered',\
            'root_path':'/ptmp/Matthew.Harrison/archive/Matthew.Harrison/ulm_mom6_d00d00100e7cc79130abeadbbda1997f698f6659/',\
            'platform':'gfdl.ncrc2-intel-prod'})
EXP.append({'name':'FLOR2_LM3_SIS2_baseline0_CHLA',\
            'root_path':'/ptmp/Matthew.Harrison/archive/Matthew.Harrison/ulm_mom6_d00d00100e7cc79130abeadbbda1997f698f6659/',\
            'platform':'gfdl.ncrc2-intel-prod'})
```

```python
sgrid=supergrid(file='ocean_hgrid.nc',tripolar_n=True,cyclic_x=True)
grid=quadmesh(supergrid=sgrid)
grid.D=nc.Dataset('topog.nc').variables['depth'][:]
grid.wet=np.zeros(grid.D.shape)
grid.wet[grid.D>0.]=1.
grid.lath=grid.y_T[:,grid.im/4]
mask=nc.Dataset('basin_codes.nc').variables['basin'][:]
AtlMask=mask.copy()
AtlMask[mask==1]=0.
AtlMask[mask==2]=1.
AtlMask[mask==4]=1.
AtlMask[mask==6]=1.
AtlMask[mask==8]=1.
AtlMask[mask==9]=1.
AtlMask[AtlMask>1]=0.
PacMask=mask.copy()
PacMask[mask!=3]=0.
PacMask[mask==3]=1.
grid.mask=AtlMask
```

```python
S=[]
for exp in EXP:
    listing=exp['root_path']+exp['name']+'/'+exp['platform']+'/history/*/*.ocean_month.nc'
    listing=glob.glob(listing)
    listing=sorted(listing)
#    print listing[0],listing[-1]
    S.append(state(MFpath=listing,grid=grid,\
                   fields=['PRCmE','net_heat_surface'],verbose=False,\
#                   date_bounds=[datetime.datetime(1992,1,2),datetime.datetime(2004,1,1)],\
                   default_calendar='gregorian'))
#    print S[-1].var_dict['PRCmE']['dates']
    S[-1].grid.mask=grid.mask
    S[-1].mask_where('PRCmE','grid.mask == 0' )
    S[-1].adjust_thickness('PRCmE')
    S[-1].volume_integral('PRCmE','X',normalize=False)
    S[-1].mask_where('net_heat_surface','grid.mask == 0' )
    S[-1].volume_integral('net_heat_surface','X',normalize=False)
```

```python
%matplotlib inline

fig=plt.figure(1,figsize=(12,6))
ax1=fig.add_subplot(111)

colList=['b','r','g','c','k'][::-1]
x=grid.lath
for res,exp in zip(S,EXP):
    col=colList.pop()
    print col,exp['name']
    res.time_avg('PRCmE_xint')
    y=res.PRCmE_xint_tav
    y=sq(np.ma.filled(y,0.))
    y=np.sum(y)-np.cumsum(y)
    ax1.plot(x,y/1.e9,color=col,label=exp['name'])

ax1.plot([x[0],x[-1]],[-0.28,-0.28],color='k',linestyle='dashed',label='Talley08')
ax1.plot([-30.,-30.],[-0.6,0.75],color='k',linestyle='dashed',)
ax1.set_xlim(-38,90)
ax1.set_ylim(-.6,.75)
plt.legend()
ax1.grid()
ax1.set_title('Implied Southward FW Transport (Sv)')
```

```python
V=[]
for exp in EXP:
    listing=exp['root_path']+exp['name']+'/'+exp['platform']+'/history/*/*.ocean_annual_z.nc'
    listing=glob.glob(listing)
    listing=sorted(listing)
    V.append(state(MFpath=listing,grid=grid,fields=['vh'],verbose=False))
    V[-1].grid.mask=grid.mask
    V[-1].adjust_thickness('vh')
    V[-1].mask_where('vh','grid.mask == 0' )
```

```python
%matplotlib inline

fig=plt.figure(1,figsize=(12,6))
ax1=fig.add_subplot(111)
x=grid.lath
indx=np.where(np.logical_and(x>40,x<50.))
colList=['b','r','g','c','k'][::-1]
for res,exp in zip(V,EXP):
    time=instance_to_datetime(res.var_dict['vh']['dates'])
    if len(time) < 2: continue
    y=res.vh
    y=np.ma.filled(y,0.)
    y=np.ma.sum(y,axis=3)
    y=np.cumsum(y,axis=1)/1.e9
    y=sq(y[:,:,indx])
    y=np.max(y,axis=2)
    y=np.max(y,axis=1)
    ax1.plot_date(time,y,linestyle='solid',color=colList.pop(),label=exp['name'])

plt.legend()
```

```python
%matplotlib inline

fig=plt.figure(1,figsize=(12,6))
ax1=fig.add_subplot(111)

expi = 3

colList=['b','r','g','c'][::-1]
x=grid.lath
y=V[expi].var_dict['vh']['zax_data']

X,Y=np.meshgrid(x,y)

res=V[expi]
res.time_avg('vh')
z=sq(np.ma.filled(res.vh_tav,0.))
z=np.ma.sum(z,axis=2)
z=np.cumsum(z,axis=0)
cf=ax1.pcolormesh(X,-Y,z/1.e9,vmin=-8,vmax=20,cmap=plt.cm.spectral)
cs=ax1.contour(X,-Y,z/1.e9,np.linspace(-30,30,15),colors='k')
plt.clabel(cs,format='%d3')
#ax1.plot([x[0],x[-1]],[-0.28,-0.28],color='k',linestyle='dashed',label='Talley08')
#ax1.plot([-30.,-30.],[-0.6,0.75],color='k',linestyle='dashed',)
ax1.set_xlim(-35,90)
ax1.set_ylim(-5500,0)
ax1.grid()
plt.colorbar(cf,ax=ax1)
tit=EXP[expi]['name']+' Atlantic Overturning Streamfunction (Sv)'
ax1.set_title(tit)
```
